# Unique SparseNet

```python
import networkx as nx
```

Let's make a sample graph.

```python
G = nx.gnm_random_graph(500, 1000)
G = G.subgraph(next(nx.connected_components(G)))
nx.draw(G, with_labels=True)
```

Let's see all the shortest paths of length diameter.

```python
diameter = nx.diameter(G)
degrees = {node:len(adj) for node, adj in G.adjacency()}
diameters = []
for src in G.nodes():
    for dest in G.nodes():
        for path in nx.all_shortest_paths(G, src, dest):
            if len(path) == diameter + 1:
                diameters.append((sum(degrees[node] ** 2 for node in path), path))
```

```python
biggest = max(a for a, b in diameters)
best = [path for val, path in diameters if val == biggest]
best
```
